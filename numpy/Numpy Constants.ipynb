# Numpy Constants

```python
import numpy as np
```

| Constants   | Meaning           |
|-------------|-------------------|
| inf         | Positive Infinity |
| NINF        | Negative Infinity |
| PZERO       | Positive Zero     |
| NZERO       | Negative Zero     |
| NAN         | Not a Number      |
| e           | Euler Value       |
| euler_gamma | Euler Gamma Value |
| pi          | PI value          |

```python
print(np.inf)
```

```python
print(np.NINF)
```

```python
print(np.PZERO)
```

```python
print(np.NZERO)
```

```python
print(np.NAN)
```

```python
print(np.e)
```

```python
print(np.euler_gamma)
```

```python
print(np.pi)
```
